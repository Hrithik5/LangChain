## Build a Simple Workflow/Graph using LangGraph

### State
- Define the state of the graph.
- The state schema serves as the input schema for all the nodes and edges in the graph.
- Using TypeDict class as our scheama, which provides types hints for the keys.

```python
from typing_extensions import TypedDict

class State(TypedDict):
    graph_info: str
```

```python
def start_play(state:State):
    print("Start play Node has be called")
    return {"graph_info" : state["graph_info"] + " I am planning to play "}

def cricket(state:State):
    print("Cricket play has be called")
    return {"graph_info": state["graph_info"] + "Cricket"}

def badminton(state:State):
    print("Cricket play has be called")
    return {"graph_info": state["graph_info"] + "Badminton"}
```

```python
import random
from typing import Literal

def random_play(state:State) -> Literal['cricket', 'badminton']:
    if random.random() > 0.5:
        return "cricket"
    else :
        return "badminton"
```

```python
from IPython.display import Image,display
from langgraph.graph import StateGraph, START, END

## Build the graph
graph = StateGraph(State)

## Add all nodes
graph.add_node("start_play", start_play)
graph.add_node("cricket", cricket)
graph.add_node("badminton", badminton)

## Schedule the flow of the graph
graph.add_edge(START, "start_play")
graph.add_conditional_edges("start_play", random_play) # Conditional edge
graph.add_edge("cricket", END)
graph.add_edge("badminton", END)

## Compile the graph
graph_builder = graph.compile()

## View 
display(Image(graph_builder.get_graph().draw_mermaid_png()))
```

```python
graph_builder.invoke({"graph_info": "My name is Hrithik "})
```
